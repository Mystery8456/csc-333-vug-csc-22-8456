```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog

# Coefficients of the objective function
c = [-3, -4]

# Coefficients of the constraints
A = [[2, 1], [1, 2]]
b = [12, 8]

# Bounds for each variable
x0_bounds = (0, None)
x1_bounds = (0, None)

# Solve the linear program
result = linprog(c, A_ub=A, b_ub=b, bounds=[x0_bounds, x1_bounds], method='highs')

print("Optimal Solution:", result.x)
print("Maximum Profit:", -result.fun)

# Plotting the constraints and feasible region
x = np.linspace(0, 6, 400)
y1 = (12 - 2*x)  # 2A + B <= 12
y2 = (8 - x) / 2  # A + 2B <= 8

plt.plot(x, y1, label=r'$2A + B \leq 12$')
plt.plot(x, y2, label=r'$A + 2B \leq 8$')
plt.fill_between(x, np.minimum(y1, y2), 0, where=(x >= 0) & (x <= 4), color='gray', alpha=0.3)

plt.xlim((0, 6))
plt.ylim((0, 6))
plt.xlabel('A')
plt.ylabel('B')
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.legend()
plt.title("Maximizing Profit")
plt.show()
```
